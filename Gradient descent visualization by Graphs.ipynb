```python
import numpy as np
import time
import matplotlib.pyplot as plt

def step_gradient(data, learning_rate, m, c):
    dcost_dm=0
    dcost_dc=0
    plt.scatter(data[:, 0], data[:, 1], alpha=0.3, c='r')
    for i in range(len(data)):
        x=data[i, 0]
        y=data[i, 1]
        dcost_dm+=(-2/len(data))*(y-m*x-c)*x
        dcost_dc+=(-2/len(data))*(y-m*x-c)
    new_m=m-learning_rate*dcost_dm
    new_c=c-learning_rate*dcost_dc
    
    x_plot=np.arange(0, 200)
    y_plot=new_m*x_plot+new_c
    plt.axis([22, 75, 20, 120])
    plt.plot(x_plot, y_plot)
    plt.show()
    
    return (new_m, new_c)

def gd(data, iterations, learning_rate):
    m=0
    c=0
    for i in range(iterations):
        m, c=step_gradient(data, learning_rate, m, c)
    return m, c

def run():
    data=np.loadtxt('datasets/data.csv', delimiter=',')
    iterations=100
    learning_rate=0.00001
    m, c=gd(data, iterations, learning_rate)
    print(m,c)
run()
```
